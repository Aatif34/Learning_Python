### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
from IPython.core.compilerop import check_linecache_ipython
# Write your code here

heads = int(input("Enter number of heads: "))
legs = int(input("Enter number of legs: "))

for dogs in range(heads + 1):

  chicken = heads - dogs

  total_legs = 4*dogs + 2*chicken

  if total_legs == legs:
    print(f"number of dogs:  {dogs}")
    print(f"number of chicken: {chicken}")

````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("enter n: "))
sum = 0
for i in range(1, n+1):
  sum += i*i

print(f"sum of squares of first n natural numbers: {sum}")
````
